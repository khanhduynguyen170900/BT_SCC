```python
import math
import datetime


def is_number(n):
    if isinstance(n, int) or isinstance(n, float):
        return True
    return False

def is_int(n):
    if isinstance(n, int):
        return True
    return False

def is_positive_int(n):
    if is_int(n)and n > 0:
        return True
    return False

def is_negative_int(n):
    if is_int(n) and n < 0:
        return True
    return False
    
"""============================================================================
1 Viết function truyền vào số nguyên dương n. Trả về list bình phương của
các số từ 1->n.
VD: 
input: 4
output: [1, 4, 9, 16]
"""

def _pow(n, p=2):
    if not is_positive_int(n):
        return 'Tham số n phải là số nguyên dương.'
    return [pow(x,p) for x in range(1, n+1)]


"""============================================================================
2. Viết function truyền vào 2 tham số m:số dân hiện tại và n:số năm
trả về dân số của 1 nước sau n năm, tì lệ tăng dân số là 1.8% hàng năm
"""

def increase_population(m, n):
    
    if not is_positive_int(m):
        return 'Tham số m phải là số nguyên dương.'
    if not is_number(n):
        return 'Tham số n phải là kiểu int hoặc float.'
    elif n < 0:
        return 'Tham số n phải là số dương.'
    
    return round(m * 1.018**n)


"""============================================================================
3. Viết function cho người dùng nhập chỉ số điện mới và cũ từ bàn phím.
    In ra màn hình chỉ số cũ, 
    chỉ số mới, 
    tiền trả định mức, 
    tiền trả vượt định mức, 
    tổng tiền phải trả. 

Định mức sử dụng điện cho mỗi hộ là: 50 KW với giá 2300đ/KW
Nếu phần vượt định mức <= 50KW thì tính giá 4800đ/KW
Nếu 50KW < phần vượt định mức < 100KW thì tính giá 7000đ/KW 
Nếu phần vượt định mức >= 100KW thì tính giá 9000đ/KW 
"""

def print_elec_info(old_number, new_number):
    
    if not (is_number(old_number) and is_number(new_number)):
        return 'Tham số phải là kiểu int hoặc float.'
    elif old_number < 0 or new_number < 0:
        return 'Tham số phải là số dương.'
    
    use_KW = new_number-old_number
    if use_KW < 0:
        print('Người dùng đã nhập sai!')
        return
    
    pay_dm = 50*2300
    
    pay_vdm = 0
    
    if use_KW >= 150:
        pay_vdm = (use_KW-50)*9000
    elif use_KW > 100:
        pay_vdm = (use_KW-50)*7000
    elif use_KW > 50:
        pay_vdm = (use_KW-50)*4800
    else:
        pay_dm = use_KW*2300
    
    print(f"Chỉ số cũ: {old_number}")
    print(f"Chỉ số mới: {new_number}")
    print(f"Tiền trả định mức: {pay_dm}")
    print(f"Tiền trả vượt định mức: {pay_vdm}")
    print(f"Tiền phải trả: {pay_vdm + pay_dm}")

    
"""============================================================================
4. Viết function trả về kết quả xếp loại học lực của học sinh. 
Giả sử học sinh chỉ có bốn cột điểm Toán (T), Lý (L), Hóa (H) và Anh văn (AV). 
Điểm 4 môn được nhập vào từ bàn phím và theo thang điểm 100. 

Qui tắc xếp loại như sau:
    + Nếu trung bình 4 môn lớn hơn hoặc bằng 80 và không có môn nào nhỏ hơn 65, 
    thì xếp loại Giỏi.
    + Nếu học sinh không đủ điều kiện đạt loại giỏi, mà có trung bình 4 môn 
    lớn hơn hoặc bằng 65 và không có môn nào nhỏ hơn 50, thì xếp loại Khá.
    + Nếu học sinh không đủ điều kiện đạt loại Khá, mà có trung bình 4 môn 
    lớn hơn hoặc bằng 50 và không có môn nào nhỏ hơn 20, thì xếp loại 
    Trung bình.
    + Nếu học sinh không đạt 3 loại trên thì xếp loại Yếu.
"""


def rank_student(T, L, H, AV):
    
    if not (is_number(T) and is_number(L) and is_number(H) and is_number(AV)):
        return 'Tham số phải là kiểu int hoặc float.'
    
    if T < 0 or L < 0 or H < 0 or AV < 0 or T > 100 or L > 100 or H > 100 or AV > 100:
        return 'Điểm một trong các môn học đã bị sai. Phải nằm trong [0,100]'
    
    dtb = (T+L+H+AV)/4
    
    if dtb >= 80 and (T >= 65 and H >= 65 and L >= 65 and AV >= 65):
        return 'Giỏi'
    elif dtb >= 65 and (T >= 50 and H >= 50 and L >= 50 and AV >= 50):
        return 'Khá'
    elif dtb >= 50 and (T >= 20 and H >= 20 and L >= 20 and AV >= 20):
        return 'Trung bình'
    return 'Yếu'


"""============================================================================
5. Viêt function truyền vào 2 số A < B < 2000, trả về danh sách chứa tất cả
các số chia hết cho 7 nhưng không chia hết cho 5, nằm trong đoạn từ A 
đến B(tính cả A và B)
"""

def ex5(A, B):
    if not (is_number(A) and is_number(B)):
        return 'Tham số phải là kiểu int hoặc float.'
    elif A < 0 or B < 0:
        return 'Tham số phải là số dương.'
    
    if B >= 2000 or A >= B:
        print("A và B không thoả yêu cầu đề bài (A < B < 2000).")
    
    a = math.ceil(A / 7)
    b = math.floor(B / 7)
    result = []
    for x in range(a, b+1):
        if x % 5 == 0:
            continue
        result.append(7*x)
    return result
    

"""============================================================================
6. Viết function in ra màn hình xem chuỗi kí tự nhập vào từ người dùng có hợp
lệ hay không. Chuỗi hợp lệ là chuỗi có chiều dài ít nhất 8 kí tự, chứa ít nhất
một chữ in hoa, chứa ít nhất một kí tự dạng số.
"""

def is_valid_str():
    s = input('Nhập vào một chuỗi hợp lệ: ')
    
    has_cap = any([c.isupper() for c in s])
    has_digit = any([c.isdigit() for c in s])
    
    if len(s) >= 8 and has_cap and has_digit:
        return True
    return False

"""============================================================================
7. Tạo danh sách các số nguyên dương nhỏ hơn 60000 có đặc điểm sau: số đó 
gồm n chữ số, tổng lũy thừa bậc n của các chữ số đó bằng chính số đó
VD: 
    153 là số có 3 chữ số, 
    và 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153
"""


def ex7(m):
    
    if not is_int(m):
        return 'Tham số phải là kiểu int.'
    if m >= 60000:
        return 'Tham số phải bé hơn 60000.'
    
    results = []
    for number in range(1, m):
        
        # Split number to digits: 157 --> [1,5,7]
        digits = [int(d) for d in str(number)]

        # Find sum
        n = len(digits)
        sum_pow_digits_n = sum([d**n for d in digits])
        
        # Check the requirement
        if sum_pow_digits_n == number:
            results.append(number)
    return results


"""============================================================================    
8. Cho trước 2 dictionary, viết chương trình kết hợp 2 dictionary đó, đối với 
các item có cùng key thì cộng value lại với nhau.
VD
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
output: {'a': 400, 'b': 400, 'd': 400, 'c': 300}
"""

def is_int_list(l):
    return [isinstance(e, int) for e in l]

def is_float_list(l):
    return [isinstance(e, float) for e in l]

def is_number_list(l):
    combine_int_float = list(zip(is_int_list(l), is_float_list(l)))
    return all([_is_int or _is_float for _is_int, _is_float in combine_int_float])


def merge_dict(d1, d2):
    # Check all value in dict is number type (int and float)
    is_number_list1 = is_number_list(list(d1.values()))
    
    is_number_list2 = is_number_list(list(d2.values()))
    
    if not (is_number_list1 and is_number_list2):
        return '1 trong 2 dictionary chứa dữ liệu không phải dạng số (int hoặc float).'

    # Vì dictionary là kiểu dữ liệu mutable nên chúng ta tạo bản copy để giá trị gốc không bị thay đổi.
    _dict1 = d1.copy() # Thêm giá trị key trùng khớp của d2 vào d1, và remove key đó trong d2.
    _dict2 = d2.copy() # Các key của d2 không tồn tại trong d1 sẽ được thêm vào sau vòng lặp for.

    keys_1 = _dict1.keys()
    keys_2 = _dict2.keys()

    for key in keys_1:
        if key in _dict2:
            # Update value on dict1[key]
            _dict1[key] = _dict1[key] + _dict2[key]

            # Remove dict2[key]
            del _dict2[key]
            
    return {**_dict1, **_dict2}
                
                
                
"""============================================================================
9. Viết function truyền vào số nguyên dương n <= 30, trả về dictionary có độ 
dài = n sao cho mỗi cặp item có dạng (x:x*x).
VD: 
với n = 5
output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
"""

def ex9(n):
    if not is_positive_int(n):
        return 'Tham số phải là số nguyên dương.'
    elif n > 30:
        return 'Tham số chỉ được bé hơn hoặc bằng 30.'
    return {x: x*x for x in range(1,n+1)}


"""============================================================================
10. Viết function tính tần suất xuất hiện các từ(cách nhau bởi dấu cách " ") 
từ input. In ra màn hình kết quả sau khi đã sắp xếp theo bảng chữ cái.
VD Input: 
New to Python or choosing between Python 2 and Python 3? Read Python 2 or 
Python 3.
- Output: 2:2 , 3.:1 , 3?:1 , New:1 , Python:5 , Read:1 , and:1 , between:1 ,
 choosing:1 , or:2 , to:1
"""

def ex10():
    s = input("Nhập vào chuỗi cần thống kê: ")
    
    list_s = s.split()
    
    dict_count = {}
    
    for w in list_s:
        if w in dict_count:
            dict_count[w] += 1
        else:
            dict_count[w] = 1
    return dict_count



"""============================================================================
11.Viết module triangle trong đó có các hàm:
- kiểm tra xem tam giác là tam giác cân, tam giác đều, tam giác vuông 
hay thường?
- tính chu vi tam giác
- tính diện tính tam giác.
input các hàm trên là chiều dài 3 cạnh tam giác.
"""


def detail_of_triangle(a, b, c):
    if not (is_number(a) and is_number(b) and is_number(c)):
        return 'Tham số phải là kiểu int hoặc float.'
    elif a <= 0 or b <= 0 or c <= 0:
        return 'Tham số phải là số dương khác không.'
    
    sides = [a,b,c]
    sides.sort()
    
    # Kiểm tra tính chất tất yếu của tam giác
    if sides[0] + sides[1] <= sides[2]:
        return 'Độ dài của 3 cạnh này không thể tạo thành một tam giác.'
    
    # Tính diện tích tam giác bằng công thức Heron (không cần biết tính chất và tìm chiều cao tam giác)
    S = 1/4 * ((a+b+c)*(a+b-c)*(a+c-b)*(b+c-a))**(1/2)
    
    result = {'Tính chất': 'Tam giác thường', 'Chu vi': a+b+c, 'Diện tích':S}
    if (a == b) or (a == c) or (b == c):
        if a == b == c:
            result['Tính chất'] = 'Tam giác đều'
        else:
            result['Tính chất'] = 'Tam giác cân'
    elif sides[2] == (sides[0]**2 + sides[1]**2)**(1/2):
        result['Tính chất'] = 'Tam giác vuông'
    return result


"""============================================================================
12. Xây dựng function trả về ma trận đơn vị kích thước N*N
VD Input: N = 3
Ouput = [[1,0,0],[0,1,0],[0,0,1]]
"""

def create_empty_square_matrix(N):
    matrix = []
    for i in range(N):
        row = [0]*N
        matrix.append(row)
    return matrix
    
def create_identity_matrix(N):
    if not is_positive_int(N):
        return 'Tham số n phải là số nguyên dương.'
    
    identity_matrix = create_empty_square_matrix(N)
    for i in range(N):
        identity_matrix[i][i] = 1
        
    return identity_matrix
    

"""============================================================================
13. Xây dựng module mymath.py gồm các function sau:
- xuất ra giá trị xuất hiện nhiều nhất trong 1 list số, trả về nhiều giá trị 
nếu có.(gọi ý dùng mode() trong statistics)
VD: Input:[1,2,5,7,3,7,7,5,8,5]
Output: 5, 7

- tính giai thừa 1 số nguyên.
- kiểm tra 1 số có phải là số nguyên tố hay không? 
(số nguyên tố là số chỉ chia hết cho 1 và chính nó.)
- kiểm tra 1 số có phải số chính phương hay không? 
(số chính phương là số có căn bậc 2 là số tự nhiên.)
Tạo 1 file main.py để gọi hàm từ module mymath
"""

from statistics import multimode

def find_mode(_list):
    if not is_number_list(_list):
        return 'Tất cả phần tử trong list phải thuộc kiểu int hoặc float.'
    return multimode(_list)

def factorial(n):
    if not is_positive_int(n):
        return 'Tham số phải là số nguyên dương.'
    return math.prod([i for i in range(1, n+1)])

def is_prime(n):
    if not is_positive_int(n):
        return 'Tham số phải là số nguyên dương.'
    if n == 1:
        return False
    
    for i in range(2, math.floor(n**(1/2))+1):
        if n % i == 0:
            return False
    return True

def is_perfect_square(n):
    if not is_positive_int(n):
        return 'Tham số phải là số nguyên dương.'
    if int(n**(1/2))**2 == n:
        return True
    return False



"""============================================================================
15.
Viết hàm nhận vào file transaction.txt với dòng đầu tiên là id của tài khoản, 
dòng thứ 2 là số tài khoản, dòng thứ 3 là số dư tài khoản, các dòng tiếp theo 
hiển thị nội dung giao dịch 
    + Với D,100 là nộp vào ngân hàng 100. 
    + Với W,100 là rút ra 100 từ tài khoản. 
Hàm trả về số dư cuối cùng của tài khoản
vd input:
Id1
Tk01
400
D,100
W,200
-> output: 300
"""

def read_transaction(filename):
    with open(filename, 'r') as f:
        content = f.readlines()
        
    content = [c.strip() for c in content if c.strip() != '']
    
    if not content[2].isdigit():
        return 'Số dư tài khoản không hợp lệ.'
    
    D = 0
    W = 0
    for row in content[3:]:
        trans = row.split(',')
        trans = [t.strip() for t in trans]
        
        if len(trans) < 2 or len(trans) > 2:
            return 'Dữ liệu sai cú pháp.'
        
        if not trans[1].isdigit():
            return 'Số tiền giao dịch của tài khoản không hợp lệ.'
        
        if int(trans[1]) <= 0:
            return 'Số tiền giao dịch của tài khoản không hợp lệ.'
        
        if trans[0] == 'D':
            D += int(trans[1])
            
        elif trans[0] == 'W':
            W += int(trans[1])
        else:
            return 'Lịch sử giao dịch của tài khoản không hợp lệ.'
        
    info_transaction = {'id': content[0], 'stk': content[1], 'sd': int(content[2]), 'D': D, 'W': W}
    
    return info_transaction['sd'] + info_transaction['D'] - info_transaction['W']


"""============================================================================
16.
File txt với nội dung file theo cú pháp sau: file gồm nhiều dòng, mỗi dòng 
tương ứng với tên học sinh và điểm số học Python được cách nhau bởi dấu phẩy.  
Ví  dụ:
Nam,10
Nữ,9

Viết hàm với input đầu vào là đường dẫn của file, hàm trả về danh sách các bạn 
được điểm cao nhất và điểm cao nhất
"""


def ex16(filename):
    with open(filename, 'r') as f:
        content = f.readlines()
        
    content = [c.strip() for c in content if c.strip() != '']
    
    students_list = {}
    _id = 0
    for row in content:
        student_info = row.split(',')
        student_info = [s.strip() for s in student_info]
        if len(student_info) < 2 or len(student_info) > 2:
            return 'Dữ liệu sai cú pháp.'
        
        if not student_info[1].isdigit():
            return 'Điểm học sinh không hợp lệ.'
        
        name = student_info[0]
        score = int(student_info[1])
        if score < 0 or score > 10:
            return 'Điểm học sinh không hợp lệ.'
        
        if name in students_list:
            students_list[name + '_' + str(_id)] = score
            _id += 1
        else:
            students_list[name] = score
            
    highest_score = max(students_list.values())
    
    result = {}
    for name in students_list.keys():
        if students_list[name] == highest_score:
            result[name] = highest_score
    return result

"""============================================================================
17.
Một Robot di chuyển trong mặt phẳng bắt đầu từ điểm đầu tiên. 
Robot có thể di chuyển theo hướng UP, DOWN, LEFT và RIGHT với những bước 
nhất định. Dấu di chuyển của robot được lấy từ file txt đầu vào, với 2 dòng 
đầu tiên thể hiện vị trí điểm đầu tiên, các dòng tiếp theo thể hiện quá trình 
di chuyển trong đó các con số phía sau hướng di chuyển chính là số bước đi 
VD: Input: 
X0 4 
Y0 3 
UP 5 
DOWN 3 
LEFT 3 
RIGHT 3 

Hãy viết hàm trả về khoảng cách từ vị trí hiện tại đến vị trí đầu tiên, 
sau khi robot đã di chuyển một quãng đường với input là tên file txt.
"""

def find_euclid_distance(filename):
    with open(filename, 'r') as f:
        content = f.readlines()
        
    content = [c.strip() for c in content if c.strip() != '']
    
    for i,row in enumerate(content):
        row = row.split()
        content[i] = [r.strip() for r in row]
        
        if len(content[i]) < 2 or len(content[i]) > 2:
            return 'Dữ liệu sai cú pháp.'
        
        if not content[i][1].isdigit():
            return 'Dữ liệu vị trí hoặc quãng đường không hợp lệ.'
    
    if content[0][0] != 'X0' or content[1][0] != 'Y0':
        return 'Dữ liệu sai cú pháp.'
    
    accept = ['UP', 'DOWN', 'LEFT', 'RIGHT']
    for row in content[2:]:
        if row[0] not in accept:
            return 'Dữ liệu sai cú pháp.'
    
    X0 = int(content[0][1])
    Y0 = int(content[1][1])
    
    X_curr = X0
    Y_curr = Y0
    
    
    dict_move = {}
    
    for row in content[2:]:
        
        direct = row[0]
        value = int(row[1])
        if direct in dict_move:
            dict_move[direct] += value
        else:
            dict_move[direct] = value
                      
    if 'UP' in dict_move:
        Y_curr += dict_move['UP']
    if 'DOWN' in dict_move:
        Y_curr -= dict_move['DOWN']
    if 'LEFT' in dict_move:
        X_curr -= dict_move['LEFT']
    if 'RIGHT' in dict_move:
        X_curr += dict_move['RIGHT']
    
    return math.sqrt((X_curr-X0)**2 + (Y_curr-Y0)**2) 
                    
                      
"""===========================================================================
18. Tìm số nguyên dương N bé nhất sao cho N chia hết cho tất cả các số trong 
đoạn [1, 20]
"""

def find_all_prime_number(start, end):
    if not (is_positive_int(start) and is_positive_int(end)):
        return 'Tham số phải là kiểu int.'
    elif start <= 0 or end <= 0:
        return 'Tham số phải là số dương khác 0.'
    
    if start > end:
        return 'Tham số start phải nhỏ hơn hoặc bằng tham số end.'
    
    # Phương pháp độ phức tạp là sqrt(end)*end
    result = list(range(2, end+1))
    
    for i in range(2, math.floor(end**(1/2))+1):
        if i in result:
            for j in range(i, math.floor(end/i) + 1):
                value_to_remove = i*j
                if value_to_remove in result:
                    result.remove(i*j)
    if start in result:
        return result[result.index(start):]
    
    result.append(start)
    result.sort()
    return result[result.index(start)+1:]
            
            

# Tìm bcnn cho một dãy số liên tục
def bcnn(start=1, end=20):
    if not (is_positive_int(start) and is_positive_int(end)):
        return 'Tham số phải là kiểu int.'
    elif start <= 0 or end <= 0:
        return 'Tham số phải là số dương khác 0.'
    
    if start > end:
        return 'Tham số start phải nhỏ hơn hoặc bằng tham số end.'
    
    prime_bcnn = find_all_prime_number(start, end)
    
    prime_bcnn_dict = dict(zip(prime_bcnn, [1]*len(prime_bcnn)))
    
    for p in prime_bcnn_dict.keys():
        prime_bcnn_dict[p] = math.floor(math.log(end, p))
        
    bcnn_value = math.prod([p**f for p, f in prime_bcnn_dict.items()])
    return bcnn_value
    

"""============================================================================
19. Viết function kiểm tra chuỗi kí tự ng dùng nhập vào từ bàn phím, điều kiện
tương tự như bài 11. Bổ sung thêm chức năng nếu như không thỏa điều kiện thì
cho phép người dùng nhập lại, nhưng tối đa ng dùng chỉ được nhập 4 lần.
"""

def ex19():
    for i in range(4):
        if is_valid_str():
            print('Chuỗi ký tự hợp lệ!')
            return True
        if i == 3:
            print(f'Bạn đã nhập sai nhiều hơn số lần cho phép!')
            return False
        print(f'Chuỗi ký tự không hợp lệ lần {i+1}. Vui lòng nhập lại!\n')
            

"""============================================================================
20.
Viết function cho phép người dùng nhập ngày, tháng, năm từ bàn phím.
In ra màn hình kiểm tra ngày người dùng vừa nhập:
    + là ngày thứ mấy trong tuần
    + là ngày thứ bao nhiêu trong năm
    + tháng đó có bao nhiêu ngày
    + năm đó có phải là năm nhuận hay không
"""

def is_nn(year):
    if year % 4 == 0 and year % 100 != 0:
        return True
    elif year % 400 == 0:
        return True
    return False

def what_day_is_it():
    day = int(input("Nhập ngày: "))
    month = int(input("Nhập tháng: "))
    year = int(input("Nhập năm: "))
    
    d = datetime.datetime(year, month, day)
    first_d = datetime.datetime(year, 1, 1)
    
    result = {'t': '', 'nbn': 0, 'tbnn': 0, 'nn': False}
    
    map_weekday = {0:'Thứ Hai', 1:'Thứ Ba', 2:'Thứ Tư', 3:'Thứ Năm', 4:'Thứ Sáu', 5:'Thứ Bảy', 6:'Chủ Nhật'}
    map_monthday = {1:31, 2:28, 3:31, 4:30, 5:31, 6:30, 7:31, 8:31, 9:30, 10:31, 11:30, 12:31}
    
    result['t'] = map_weekday[d.weekday()]
    result['nbn'] = int((d.timestamp() - first_d.timestamp())/86400 + 1)
    result['tbnn'] = map_monthday[month]
    
    if is_nn(year):
        result['nn'] = True
        if month == 2:
            result['tbnn'] = 29
    
    print(f"\n\nNgày này là {result['t']}.")
    print(f"Ngày này là ngày thứ {result['nbn']} trong năm.")
    print(f"Tháng này có {result['tbnn']} ngày.")
    if result['nn']:
        print('Năm này là năm nhuận.')
    else:
        print('Năm này không phải năm nhuận.')
    
    return result
```

```python
'''
Ex.8) load các data sau và thực hiện theo yêu cầu:
week1 = pd.read_csv(data_path + "Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(data_path + "Restaurant - Week 2 Sales.csv")
customers = pd.read_csv(data_path + "Restaurant - Customers.csv")
foods = pd.read_csv(data_path + "Restaurant - Foods.csv")

1) Join all data for analysis:
    - Concat data week1 and week2
    - Merge with data foods & customers
2) Statistic & visualize features of the data

3) Khảo sát 
  - Tỉ lệ nam nữ
  - Food theo giới tính
  - Giá theo giới tính
  - Company:
    + Tỉ lệ nam nữ đi mua hàng ở mỗi cty
    + Loại thức ăn mà các thành viên ở các cty hay mua
    + Có mua lặp lại không
    + Thống kê giá sản phẩm được mua bởi thành viên của các công ty
'''
```

```python
data_path = "Data/"

week1 = pd.read_csv(data_path + "Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(data_path + "Restaurant - Week 2 Sales.csv")
customers = pd.read_csv(data_path + "Restaurant - Customers.csv")
foods = pd.read_csv(data_path + "Restaurant - Foods.csv")
```

```python
# Concat data week1 and week2

week = pd.concat([week1, week2], ignore_index=True)
week
```

```python
week_freq = week.groupby(['Customer ID', 'Food ID']).size().reset_index()
week_freq = week_freq.rename(columns={0: 'freq'})
week_freq
```

```python
# Merge with data foods & customers NO_DUPLICATE

df_8 = pd.merge(week_freq, customers, left_on='Customer ID', right_on='ID')
df_8 = pd.merge(df_8, foods)
df_8 = df_8.drop('ID', axis=1)
df_8
```

```python
# Merge with data foods & customers DUPLICATE

df_8_dup = pd.merge(week, customers, left_on='Customer ID', right_on='ID')
df_8_dup = pd.merge(df_8_dup, foods)
df_8_dup = df_8_dup.drop('ID', axis=1)
df_8_dup
```
